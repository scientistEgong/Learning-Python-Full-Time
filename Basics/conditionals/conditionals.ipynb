Conditionals are written in codes to check if the a program meets a requirement or a condition.

Conditionals often take the form of comparison betweeen two values, checking if it is true.

Basic comparison operators.
These are also called relational operators; they are used to check the how two values relate to each other where the condition is either True or False.
------------------------------------------
Boolean comparator |  meaning
------------------------------------------
>                  | greater than
<                  | Less than
>=                 | Greater or equal to
<=                 | Less than or equal to
==                 | Not equal to
==                 | Equal to


**Boolean** are the main blocks for which conditionals in python operate, leveraging either True or False in a program.

This was named after Mathematician George Boole.

```python
# Conditionals and boolean check.
a = 2
b = 4
a == a
```

```python
a == b
```

```python
a < b
```

```python
a > b
```

**Logical Operators**
The boolean operator always takes two boolean expressions, so they are considered binary. Thus both evaluates expression if they are True.

The and Keyword

The and key word is a boolean check that evaluates if two conditions are True.

```python
# And examples.
a = 2
b = 2
a == 2 and b == a
```

```python
a == 4 and b == 2
```

The or Keyword
The or keyword sometimes used as an exclusive could mean, if one of two conditions is True.

or if the both conditions is True, such is the inclusive.

```python
# Exclusive
1 < 2 or 3 < 4
```

```python
# Inclusive
2 < 3 or 4 > 6
```

The not keyword
Just like negation in conditionals, it is used to check if a condition is not True.

```python
not True == False
```

The if statement.
The if statement executes a portion of a code only if the condition is met.

The else keyword.
The else keyword executes a portion of a code if the if condition is false.

The elif (else if) condition.
The elif keyword is executed if the first condition is False as well as the else condition being False. Thus the elif condition is executed.

Nested if conditions.
Nested if conditions is executed if the first if condition is executed for a different reason.
