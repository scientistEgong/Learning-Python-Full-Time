```python
# Alien color
"""Conditions are:
if the alien color is green, the player earns 5 points.
if alien color is yellow, the player earns 10 points.
if the alien color is red, the player earns 15 points."""

alien_color = input("Guess an Alien color and earn points: ").lower()
if alien_color == "red":
    points = 15
    print(f"You earned {points} points.")
elif alien_color == "yellow":
    points = 10
    print(f"You earned {points} points.")
elif alien_color == "green":
    points = 5
    print(f"You earned {points} points.")

else:
    print("Invalid color".)


            
```

```python
while True:
    number = int(input("Enter a positive number and 0 to quit: "))
    if number == 0:
        print("Exiting.")
        continue

    for i in range(100):
        if number % i == 0:
            print(f"{i} is a factor of {number}")

            
```

```python
while True:
    number = int(input("Enter a positive number and 0 to quit: "))
    if number == 0:
        print("Exiting.")
        break

    for i in range(1, 100):
        if number % i == 0:
            print(f"{i} is a factor of {number}")

            
```

```python
sums_of_even = 0
for n in range(11):
    if n % 2 == 0:
        sums_of_even += n
print(sums_of_even)
```
