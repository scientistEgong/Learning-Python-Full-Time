**Understanding Loops**
Loops are conditionals more like iterators that run repeatedly until a condition is met or a condition is True.

**while Loop**
The while loop is an iterator or a statement, block or body of code that runs repeatedly while a condition is True and it faulters when the same condition is false.

It follows this convention;
- The while statement.
- The loop body

The while loop is quite similar to the if statement but with different perculiarities.

The break statement is used to end the iteration of a loop, either it is a for loop or it is a while loop.

The continue statements evaluates a condition under a loop body when the condition is True, it skips the current iteration and moves to the next iteration.

```python
# Testing basic examples on while loop.
current_number = 9
while current_number > 0:
    current_number -= 1
    print(current_number)
```

```python
number = int(input("Enter a number: "))
conditions = (number % 2 == 0)
while conditions:
    print("Number is an even number")
    break

else:
    print("It is not an even digit.")    
```

**The for loops**
This executes a section of a code once for each item in a collection of items.
The for loop follows this sequence;
- The for statement
- The loop body.

Using the range function.
The range method is used to list out all numbers between two numbers.


The enumerate function also seperates characters in a string or collections to indices.

Nested for loops
These are for loops under for loops.


```python
for n in range(1, 4):
    for j in range(4, 7):
        print(n, j)
```

```python
name_input = input("Enter a word:")
for id, names in enumerate(name_input, start=0):
    print(id, names)
```
