**Functions**
Functions are programs that perform a specific task, they can be seen as block of codes that serves as solutions to a coding challenge. Functions help in code recycling as codes serve the purpose of reusabaility.

**Types of Functions**
There are two types of functions basically.
- Implicite functions
  Implicite functions are functions that exist primarily in python without externally creating it. These type of functions are built-in and cannot be used to name variables.
  
  Examples are; 
  print()
  len()
  etc.

- Explicite Functions
  Like the name, explicite functions are functions that are created by the programmer.
  It involves defining the function, using a doc string to explain the intent of the function, the block of code, and the function call, which is the same as the function name with parentheses.


In explicite functions, there are terms that are commonly used.
- Arguments.
  Arguments are values that are passed to the functions as inputs.
  They are usually passed to the function call.

- Return
  The return statement returns the value as output, depending on the scenario. If the funtion call is not printed, it returns what is stored in the funtion but it doesn't display it. If it's printed it returns and displays the output to the user.

- def keyword
  Functions usually start with the def keyword, which is short for define.

- Parameters
  parameters are variables that represent the values of which the functions interact with when called.
  Parameters are variables that contain arguments.      
 


```python
def greet():  # Function definition
    """write a basic script that greets the user.
    This is a doc string and it explains the intent
    of the code."""
    return "Good morning! It's great to have you around."  # Function body.

print(greet())  # Function call
```

```python
import random as rn

def number_guess():
    """Guess from a random number."""
    play = True

    while play:
        raw_input = input("Type in 3 numbers separated by commas or (q to quit): ")

        # Exit code
        if raw_input.lower() == "q":
            print("Exiting...")
            break

        # Normalize input
        cleaned_input = raw_input.replace(" ", "")
        comma_split = cleaned_input.split(",")

        # Validation
        if len(comma_split) != 3:
            print("Please enter exactly three numbers.")
            continue

        if not all(dgt.isdigit() for dgt in comma_split):
            print("Only numbers are allowed.")
            continue

        # User input → [[1], [2], [3]]
        list_container = [[int(dgt)] for dgt in comma_split]

        # Random number → [n]
        random_check = [rn.randint(1, 10)]

        # Comparison (list-to-list)
        if random_check in list_container:
            print("You guessed correctly!")
        else:
            print("Wrong guess.")

        print(list_container, "\n=>\n", random_check)

number_guess()
```

```python
import random as rn

def number_guess():
    """Guess from a random number."""
    while True:
        raw_input = input("Type in 3 numbers separated by commas or (q to quit): ")

        if raw_input.lower() == "q":
            print("Exiting...")
            break

        # Conditionals
        cleaned_input = raw_input.replace(" ", "")
        comma_split = cleaned_input.split(",")

        # Validation
        if len(comma_split) != 3:
            print("Please enter exactly three numbers.")
            continue

        if not all(dgt.isdigit() for dgt in comma_split):
            print("Only numbers are allowed.")
            continue

        # User numbers as [[1], [2], [3]]
        list_container = [[int(dgt)] for dgt in comma_split]

        # Random number as [n]
        random_check = [rn.randint(1, 10)]

        # Matching sublists
        match = [x for x in list_container if x == random_check]

        if match:
            print("You guessed correctly!")
        else:
            print("Wrong guess!")

        print("Your numbers:", list_container)
        print("Matches:", match)
        print("Random number:", random_check, "\n")

number_guess()
```
